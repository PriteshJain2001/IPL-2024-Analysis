# Note on Abandoned Matches
The matches on May 13, 2024 (Gujarat vs Kolkata), May 16, 2024 (Hyderabad vs Gujarat), and May 19, 2024 (Rajasthan Royals vs Kolkata Knight Riders) were abandoned due to various reasons, including rain.
In these cases, no "Player of the Match" was awarded, and the player_of_the_match column contains the value "None".
Additionally, the most_runs and most_wkts columns also contain the value "None", reflecting the fact that no player statistics were recorded.
These "None" values are excluded from the player performance analysis, as no player was chosen or awarded in these matches.

# Note on Abandoned Matches

```python
batting_first = ipl[(ipl['first_score'] > ipl['second_score']) & (ipl['winner'] == ipl['team1'])]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(7, 7))
plt.hist(batting_first['first_score'] - batting_first['second_score'], bins=20, edgecolor='black', color='#FF6347')  
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Histogram of Wins by Runs (Batting First)', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Wins by Runs', fontsize=12, color='#FFFFFF')
plt.ylabel('Frequency', fontsize=12, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
player_of_match_count = ipl['player_of_the_match'].value_counts()

most_valuable_player = player_of_match_count.idxmax()
most_valuable_player_awards = player_of_match_count.max()

print(f'The Most Valuable Player is {most_valuable_player} with {most_valuable_player_awards} "Player of the Match" awards.')
```

```python
high_run_chases = ipl[ipl['second_score'] > ipl['first_score']]
top_performers_chase = high_run_chases['most_runs'].value_counts().reset_index()
top_performer = top_performers_chase.head(1)
print(f"Top performer in high-run chases: {top_performer}")
```

```python
ipl['winning_margin'] = ipl.apply(lambda row: row['first_score'] - row['second_score'] if row['winner'] == row['team1'] else row['second_score'] - row['first_score'], axis=1)
average_margin = ipl.groupby('winner')['winning_margin'].mean().reset_index()

print(f"Average winning margins for each team: \n{average_margin}")
```

```python
team1_avg_score = ipl.groupby('team1')['first_score'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='team1', y='first_score', data=team1_avg_score, palette='Set1')
plt.gcf().set_facecolor('#1E3A5F') 
plt.gca().set_facecolor('#2A4D69')  
plt.title('Average Score for Teams Batting First in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Team', fontsize=12, color='#FFFFFF')
plt.ylabel('Average Score', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
player_awards = ipl['player_of_the_match'].value_counts().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='index', y='player_of_the_match', data=player_awards.head(10), palette='Set2')
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Top 10 Players with Most Player of the Match Awards in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Player', fontsize=12, color='#FFFFFF')
plt.ylabel('Number of Awards', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
# Create a new column to differentiate home vs away matches
ipl['home_match'] = ipl.apply(lambda row: 1 if row['team1'] == row['winner'] else 0, axis=1)
home_win_percent = ipl.groupby('team1')['home_match'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='team1', y='home_match', data=home_win_percent, palette='Set3')
plt.gcf().set_facecolor('#1E3A5F') 
plt.gca().set_facecolor('#2A4D69')  
plt.title('Home Win Percentage for Each Team in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Team', fontsize=12, color='#FFFFFF')
plt.ylabel('Home Win Percentage', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
team_performance = ipl.pivot_table(index='team1', values=['first_score', 'first_wkts'], aggfunc='mean')
plt.figure(figsize=(10, 7))
sns.heatmap(team_performance, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='white')
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Team Performance Heatmap (Runs and Wickets) in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Performance Metrics', fontsize=12, color='#FFFFFF')
plt.ylabel('Teams', fontsize=12, color='#FFFFFF')
plt.show()
```

```python
matches_played = ipl['team1'].value_counts()
win_percentages = ipl.groupby('team1')['winner'].apply(lambda x: (x == x.name).mean() * 100).reset_index(name='win_percent')
plt.figure(figsize=(10,6))
sns.barplot(x='team1', y='win_percent', data=win_percentages, palette='Set2')
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Team Win Percentages in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Teams', fontsize=12, color='#FFFFFF')
plt.ylabel('Win Percentage', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
team_performance = ipl.groupby('team1').agg({
    'first_score': 'mean',
    'first_wkts': 'mean'
}).reset_index()
plt.figure(figsize=(10,6))
team_performance.set_index('team1').plot(kind='bar', figsize=(10,6), color=['#FFD700', '#1E90FF'], edgecolor='white')
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Average Runs and Wickets per Team in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Teams', fontsize=12, color='#FFFFFF')
plt.ylabel('Average', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
top_matches = ipl[['team1', 'team2', 'first_score', 'second_score']].sort_values(by=['first_score', 'second_score'], ascending=False).head(5)
plt.figure(figsize=(10,6))
top_matches.set_index('team1')[['first_score', 'second_score']].plot(kind='bar', stacked=True, color=['#FFD700', '#1E90FF'], edgecolor='white')
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Top 5 Highest Scoring Matches in IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Teams', fontsize=12, color='#FFFFFF')
plt.ylabel('Scores', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.show()
```

```python
top_run_scorers = ipl['most_runs'].value_counts().head(5)
top_wicket_takers = ipl['most_wkts'].value_counts().head(5)
plt.figure(figsize=(10, 6))
top_run_scorers.plot(kind='bar', color='#FF6347', label='Top Run-Scorers', edgecolor='white', alpha=0.8)
top_wicket_takers.plot(kind='bar', color='#4682B4', label='Top Wicket-Takers', edgecolor='white', alpha=0.8)
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Dominating Performers: Top Run-Scorers and Wicket-Takers of IPL 2024', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Players', fontsize=12, color='#FFFFFF')
plt.ylabel('Count', fontsize=12, color='#FFFFFF')
plt.xticks(rotation=45, color='#FFFFFF')
plt.grid(axis='y', linestyle='--', color='#B0C4DE', alpha=0.7)  
plt.legend(loc='upper right', fontsize=12, frameon=False, labelcolor='white')
plt.show()
```

```python
ipl['date'] = pd.to_datetime(ipl['date'])  
toss_over_time = ipl.groupby([ipl['date'].dt.month, 'decision'])['winner'].count().unstack().fillna(0)
plt.figure(figsize=(10, 6))
toss_over_time.plot(kind='line', marker='o', linewidth=2, markersize=8, color=["#FFD700", "#1E90FF"], ax=plt.gca())
plt.gcf().set_facecolor('#1E3A5F')  
plt.gca().set_facecolor('#2A4D69')  
plt.title('Wins Over Time Based on Toss Decision (Bat/Field)', fontsize=14, color='#FFFFFF', weight='bold')
plt.xlabel('Month', fontsize=12, color='#FFFFFF')
plt.ylabel('Number of Wins', fontsize=12, color='#FFFFFF')
plt.legend(title="Toss Decision", frameon=False, loc="upper right")
plt.grid(True, color='#B0C4DE', linestyle='--', alpha=0.7)  
plt.show()
```
